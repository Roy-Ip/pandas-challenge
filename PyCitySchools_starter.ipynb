```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Grouping the dataset by "school_name" 
grouped_df = school_data_complete.groupby(["school_name"])


# Extracting the school types and creating a dataframe
school_summary_school_type_df = grouped_df["type"].first()
school_summary_school_type_df = pd.DataFrame(school_summary_school_type_df)
school_summary_school_type_df = school_summary_school_type_df.rename(columns={"type" : "School Type"})


# Extracting the total number of students and creating a dataframe
school_summary_total_students_df = grouped_df["size"].first()
school_summary_total_students_df = pd.DataFrame(school_summary_total_students_df)
school_summary_total_students_df = school_summary_total_students_df.rename(columns={"size" : "Total Students"})


# Extracting the total budget for each school and creating a dataframe
school_summary_total_budget_df = grouped_df["budget"].first()
school_summary_total_budget_df = pd.DataFrame(school_summary_total_budget_df)
school_summary_total_budget_df = school_summary_total_budget_df.rename(columns={"budget" : "Total School Budget"})


# Calculating per student budget and creating a dataframe
school_summary_per_student_budget_df = school_summary_total_budget_df["Total School Budget"] / school_summary_total_students_df["Total Students"]
school_summary_per_student_budget_df = pd.DataFrame(school_summary_per_student_budget_df , columns=["Per Student Budget"])


# Calculating average maths scores for each school and creating a dataframe
school_summary_average_maths_score_df = grouped_df["maths_score"].mean()
school_summary_average_maths_score_df = pd.DataFrame(school_summary_average_maths_score_df)
school_summary_average_maths_score_df = school_summary_average_maths_score_df.rename(columns={"maths_score" : "Average Maths Score"})


# Calculating average reading scores for each school and creating a dataframe
school_summary_average_reading_score_df = grouped_df["reading_score"].mean()
school_summary_average_reading_score_df = pd.DataFrame(school_summary_average_reading_score_df)
school_summary_average_reading_score_df = school_summary_average_reading_score_df.rename(columns={"reading_score" : "Average Reading Score"})


# Calculating the number of students passing maths for each school and creating a dataframe
school_summary_passing_math_count_df = grouped_df["maths_score"].apply(lambda x: (x >= 50).sum())
school_summary_total_students_per_group_df = grouped_df.size()
school_summary_percent_passing_maths_df = (school_summary_passing_math_count_df / school_summary_total_students_per_group_df) * 100
school_summary_percent_passing_maths_df = pd.DataFrame(school_summary_percent_passing_maths_df , columns=["% Passing Maths"])


# Calculating the number of students passing reading for each school and creating a dataframe
school_summary_passing_reading_count_df = grouped_df["reading_score"].apply(lambda x: (x >= 50).sum())
school_summary_total_students_per_group_df = grouped_df.size()
school_summary_percent_passing_reading_df = (school_summary_passing_reading_count_df / school_summary_total_students_per_group_df) * 100
school_summary_percent_passing_reading_df = pd.DataFrame(school_summary_percent_passing_reading_df , columns=["% Passing Reading"])


# Calculating the number of students passing both maths and reading for each school and creating a dataframe
school_summary_overall_passing_count_df = grouped_df.apply(lambda x: ((x["maths_score"] >= 50) & (x["reading_score"] >= 50)).sum())
school_summary_total_students_per_group_df = grouped_df.size()
school_summary_percent_overall_passing_df = (school_summary_overall_passing_count_df / school_summary_total_students_per_group_df) * 100
school_summary_percent_overall_passing_df = pd.DataFrame(school_summary_percent_overall_passing_df , columns=["% Overall Passing"])


# Merging all the above dataframes into a single school summary dataframe
merge_df = school_summary_school_type_df.merge(school_summary_total_students_df , on = "school_name")\
                                        .merge(school_summary_total_budget_df , on = "school_name")\
                                        .merge(school_summary_per_student_budget_df , on = "school_name")\
                                        .merge(school_summary_average_maths_score_df , on = "school_name")\
                                        .merge(school_summary_average_reading_score_df , on = "school_name")\
                                        .merge(school_summary_percent_passing_maths_df , on = "school_name")\
                                        .merge(school_summary_percent_passing_reading_df , on = "school_name")\
                                        .merge(school_summary_percent_overall_passing_df , on = "school_name")

# Renaming the index name from "school_name" to  "School Name" for the dataframe for better presentation
merge_df.index.name = "School Name"


# Creating a copy of the merged dataframe for formatting purposes
school_summary = merge_df.copy()

# Formatting columns in the school summary dataframe for better presentation
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)

# Displaying the school summary dataframe
school_summary
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Grouping the 'school_data_complete' dataset by "school_name" and "year", calculating the mean maths score for each group, and resetting the index
maths_scores_by_year_groupby_df = school_data_complete.groupby(["school_name", "year"])["maths_score"].mean().reset_index()

# Grouping the previously grouped data by "school_name" and "year", calculating the mean maths score for each group, and creating a DataFrame
maths_scores_by_year = pd.DataFrame(maths_scores_by_year_groupby_df.groupby(["school_name", "year"])["maths_score"].mean().unstack())

# Setting the index name and column name
maths_scores_by_year.index.name = "School Name"
maths_scores_by_year.columns.name = " "

# Renaming the column headers of the DataFrame, otherwise it will be showing "9 , 10, 11 and 12", missing "Year" at the front
rename_column_headers_df = ["Year 9", "Year 10", "Year 11", "Year 12"] 
maths_scores_by_year.columns = rename_column_headers_df

# Displaying the dataframe showing the average maths score for students of each year level at each school
maths_scores_by_year
```

```python
# Creating a copy of "maths_scores_by_year" dataframe for formatting purpose
maths_scores_by_year_reformat = maths_scores_by_year.copy()

# Formatting the scores rounded to 2 decimal places
maths_scores_by_year_reformat["Year 9"] = maths_scores_by_year_reformat["Year 9"].map("{:.2f}".format)
maths_scores_by_year_reformat["Year 10"] = maths_scores_by_year_reformat["Year 10"].map("{:.2f}".format)
maths_scores_by_year_reformat["Year 11"] = maths_scores_by_year_reformat["Year 11"].map("{:.2f}".format)
maths_scores_by_year_reformat["Year 12"] = maths_scores_by_year_reformat["Year 12"].map("{:.2f}".format)

# Displaying the reformatted dataframe
maths_scores_by_year_reformat
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Creating the bins and labels for the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Creating a copy of "merge_df" dataframe prepared previously to avoid changing the original "merge_df" dataframe accidentally
spending_merge_df = merge_df.copy()

# Slicing the data and place it into bins
spending_merge_df["Spending Ranges (Per Student)"] = pd.cut(spending_merge_df["Per Student Budget"], spending_bins, labels=labels)

# Creating a copy of "spending_merged_df" dataframe
school_spending_df = spending_merge_df.copy()

# Grouping the data by "Spending Ranges (Per Student)" and calculating the mean for "Average Maths Score", "Average Reading Score", "% Passing Maths", "% Passing Reading" and "% Overall Passing"
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

# Creating dataframes for the above results 
spending_math_scores = pd.DataFrame(spending_math_scores)
spending_reading_scores = pd.DataFrame(spending_reading_scores)
spending_passing_math = pd.DataFrame(spending_passing_math)
spending_passing_reading = pd.DataFrame(spending_passing_reading)
overall_passing_spending = pd.DataFrame(overall_passing_spending)

# Merging the above dataframes into a single spending summary dataframe 
spending_summary = spending_math_scores.merge(spending_reading_scores , on = "Spending Ranges (Per Student)")\
                                       .merge(spending_passing_math , on = "Spending Ranges (Per Student)")\
                                       .merge(spending_passing_reading , on = "Spending Ranges (Per Student)")\
                                       .merge(overall_passing_spending , on = "Spending Ranges (Per Student)")  

# Displaying the spending summary dataframe showing the breakdown school performances based on average Spending Range (Per Student)
spending_summary
```

```python
# Creating a copy of "spending_summary" dataframe for formatting purpose
spending_summary_reformat = spending_summary.copy()

# Formatting the scores rounded to 2 decimal places
spending_summary_reformat["Average Maths Score"] = spending_summary_reformat["Average Maths Score"].map("{:.2f}".format)
spending_summary_reformat["Average Reading Score"] = spending_summary_reformat["Average Reading Score"].map("{:.2f}".format)
spending_summary_reformat["% Passing Maths"] = spending_summary_reformat["% Passing Maths"].map("{:.2f}".format)
spending_summary_reformat["% Passing Reading"] = spending_summary_reformat["% Passing Reading"].map("{:.2f}".format)
spending_summary_reformat["% Overall Passing"] = spending_summary_reformat["% Overall Passing"].map("{:.2f}".format)

# Displaying the reformatted dataframe
spending_summary_reformat
```
